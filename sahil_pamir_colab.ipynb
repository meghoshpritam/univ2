```python
# Jensen's work for software development
import matplotlib.pyplot as plt

effort = int(input("Enter effort(developers) need to develop the software: "))
time = int(input("Enter estimated time for develop a software(in months): "))
tc = [2,8,11]

def relay_graph():
  print("\n")
  
  plt.legend(loc="upper right")
  plt.xlabel("time") 
  plt.ylabel("product size")
  plt.title("r_curve for jensen")
  plt.show()

def jensen(tc,effort,time): 
  return tc * time * pow(effort, 1/2)

colors = {
      2: 'red',
      8: 'orange',
      11: 'green'
  }

points = []
k = []
for records in tc:
  a = []
  kloc = jensen(records,effort,time)
  project_duration = 3 * int(time)
  for timeTaken in range(0, project_duration):
    a.append(round(jensen(records, effort, timeTaken), 5))
  points.append(a)
  k.append(kloc)
  plt.plot(a,color=colors[records], linewidth=4, label=f"TC: {records} | KLOC: {kloc}")
  relay_graph()
i = 0
for records in tc:
  plt.plot(points[i], color=colors[records], linewidth=4, label=f"TC: {records} | KLOC: {k[i]}")
  i += 1
relay_graph()
```

```python
screensReports = []
consts = ['Enter no of views(Less then 3 then type  - 1/Between 3 - 7 then type - 2/ Greater then 7 then type - 3): ','Enter no of sections(Less then 0 - 1 then type  - 1/Between 2 - 3 then type - 2/ Greater then 4 then type - 3): ']
opPoint = [[1,2,3],[2,5,8]]
values = []
OBPOINTList = []
SumOfObPoint = 0
NOP = 0
PercentageReuse = int(input("Enter value of percentage reuse: "))
for c in consts:
  print(c)
  values.append(int(input("")))

noOfServers = int(input("Enter no of servers: \n"))
noOfClients = int(input("Enter no of clients: \n"))

total = noOfServers + noOfClients

for i in range(0,2):
  if values[i] == 1:
    if total < 8:
      screensReports.append('s')
    else:
      screensReports.append('m')
  elif values[i] == 2:
    if total < 4:
      screensReports.append('s')
    elif total < 8:
      screensReports.append('m')
    else:
      screensReports.append('d')
  elif values[i] == 3:
    if total < 4:
      screensReports.append('m')
    else:
      screensReports.append('d')

for i in range(0,2):
  if screensReports[i] == 's':
    OBPOINTList.append(opPoint[i][0])

  elif screensReports[i] == 'm':
    OBPOINTList.append(opPoint[i][1])

  elif screensReports[i] == 'd':
    OBPOINTList.append(opPoint[i][2])

OBPOINTList.append(10)
for value in OBPOINTList:
  SumOfObPoint += value

NOP = (SumOfObPoint * (100 - PercentageReuse)) / 100

capabilityRating = [4,7,13,25,50]
print("Very Low- Press 0\nLow - Press 1\nNominal - Press 2\nHigh - Press 3\nVery High - Press 4")

flag = int(input("Rate Developer experience and capability: "))
PROD = capabilityRating[flag]

effort = NOP / PROD

print("Effort need:" + str(effort))
```
